```python
import pandas as pd
from geopy.geocoders import GoogleV3
```

```python
# Function to get location details (latitude, longitude) using Google Maps Geocoding API
def get_location_details_google(address):
    geolocator = GoogleV3(api_key='YOUR_GOOGLE_MAPS_API_KEY')  # Replace with your API key
    location = geolocator.geocode(address)
    if location:
        return location.latitude, location.longitude
    else:
        return None, None
```
